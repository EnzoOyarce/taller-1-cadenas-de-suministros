```python
#a) Sin pesos
#comenzamos ordenando las coordenadas X de menor a mayor
df_x = df[['x']].sort_values(by='x').reset_index(drop=True)
df_x
# Obtener los dos valores centrales del DataFrame ordenado
mediana_x1 = len(df_x) // 2 - 1  # La posición del valor más bajo del medio
mediana_x2 = mediana_x1 + 1  # La posición del valor más alto del medio

# Seleccionar las dos filas que contienen los valores medianos de 'x'
median_values_x = df_x.iloc[[mediana_x1, mediana_x2]]

# Seleccionar las dos filas que contienen los valores medianos de 'x'
median_values_real_x = df_x.iloc[[mediana_x1, mediana_x2]].values.flatten().tolist()

# Creamos el intervalo abierto con los números obtenidos
intervalo_abierto_x = f"({median_values_real_x[0]}, {median_values_real_x[1]})"

print('La ubicacion de X optimo se encontrara en el rango de:', intervalo_abierto_x)

#Despues ordenamos las coordenadas y de menor a mayor
df_y = df[['y']].sort_values(by='y').reset_index(drop=True)
df_y
# Obtener los dos valores centrales del DataFrame ordenado considerando que son par
mediana_y1 = len(df_y) // 2 - 1  # La posición del valor más bajo del medio
mediana_y2 = mediana_y1 + 1  # La posición del valor más alto del medio

# Seleccionar las dos filas que contienen los valores medianos de 'y'
median_values_y = df_y.iloc[[mediana_y1, mediana_y2]]

# Seleccionar las dos filas que contienen los valores medianos de 'y'
median_values_real_y = df_y.iloc[[mediana_y1, mediana_y2]].values.flatten().tolist()

# Creamos el intervalo abierto con los números obtenidos
intervalo_abierto_y = f"({median_values_real_y[0]}, {median_values_real_y[1]})"

print('La ubicacion de Y optimo se encontrara en el rango de:', intervalo_abierto_y)

```

```python
#b) Con pesos
# Ordenamos el DataFrame por la columna 'x_peso' de menor a mayor
df_x_peso = df[['x', 'peso']].sort_values(by='x').reset_index(drop=True)

# Multiplicar 'x' y 'peso'
df_x_peso['producto'] = df_x_peso['x'] * df_x_peso['peso']

# Calcular la suma acumulativa y asignarla a una nueva columna llamada 'suma_acumulativa'
df_x_peso['suma_acumulativa'] = df_x_peso['producto'].cumsum()
df_x_peso
```

```python

# Ordenamos el DataFrame por la columna 'x' de menor a mayor
df_x_peso = df[['x', 'peso']].sort_values(by='x').reset_index(drop=True)

# Multiplicar 'x' y 'peso'
df_x_peso['producto'] = df_x_peso['x'] * df_x_peso['peso']

# Calcular la suma acumulativa y asignarla a una nueva columna llamada 'suma_acumulativa'
df_x_peso['suma_acumulativa'] = df_x_peso['producto'].cumsum()

# Encontrar la ubicación del X óptimo
suma_acumulativa_total = df_x_peso['suma_acumulativa'].iloc[-1]
ubicacion_optima = suma_acumulativa_total / 2

# Buscar en la tabla el primer lugar donde la suma acumulativa sea mayor o igual a la ubicación óptima
indice_optimo = df_x_peso['suma_acumulativa'].searchsorted(ubicacion_optima)
coordenada_x_optima = df_x_peso.iloc[indice_optimo]['x'] if indice_optimo < len(df_x_peso) else None

print('La coordenada X optima ocurren en: ', coordenada_x_optima)

# Ordenamos el DataFrame por la columna 'y' de menor a mayor
df_y_peso = df[['y', 'peso']].sort_values(by='y').reset_index(drop=True)

# Multiplicar 'x' y 'peso'
df_y_peso['productoy'] = df_y_peso['y'] * df_y_peso['peso']

# Calcular la suma acumulativa y asignarla a una nueva columna llamada 'suma_acumulativa'
df_y_peso['suma_acumulativay'] = df_y_peso['productoy'].cumsum()

# Encontrar la ubicación del X óptimo
suma_acumulativa_totaly = df_y_peso['suma_acumulativay'].iloc[-1]
ubicacion_optimay = suma_acumulativa_totaly / 2

# Buscar en la tabla el primer lugar donde la suma acumulativa sea mayor o igual a la ubicación óptima
indice_optimoy = df_y_peso['suma_acumulativay'].searchsorted(ubicacion_optimay)
coordenada_y_optima = df_y_peso.iloc[indice_optimoy]['y'] if indice_optimoy < len(df_y_peso) else None

print('La coordenada Y optima ocurren en: ', coordenada_y_optima)
df_y_peso
```
